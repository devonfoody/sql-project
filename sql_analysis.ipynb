5.Plusminus is an extremely important statistic in basketball to dereminte how effective someone is when they are on the court. This will be a great indicator that I will use to find out who the statistically most efficient basketball player is. I want to find the players with the best plus minus that scored over 20 points.

```python
%%sql 
SELECT Player_name,
Team_city,
PLUS_MINUS
FROM game_details
WHERE PTS > 20 
ORDER BY PLUS_MINUS DESC
LIMIT 25;
```

Here is the list of the top 25 plus minus ratios from all the data of games I have collected. As you can see here this is also a very elite list of players who scored over 20 point and positvley affected their teams on both sides of the court. 

In order to help solve my business question of who the most efficient basketball player is I decided that I would make a data request to show all players performances based on PlusMinus. This woult help me understand how on any given night they impacted their team in either a positive or negative way. 

```python
%%sql
SELECT DISTINCT player.Player_id,
player.Player_name,
game_details.Team_city,
game_details.PLUS_MINUS, 
CASE 
	WHEN PLUS_MINUS > 0 THEN 'good for team'
	WHEN PLUS_MINUS = 0 THEN 'nuetral'
	ELSE 'bad for team'
END AS overall_impact
FROM game_details
JOIN player
	ON player.Player_id = game_details.Player_id
WHERE PLUS_MINUS IS NOT NULL;
```

In this data request I created a view statement to help me see 1 home victory for every team that I have in my database. I wanted to see how many points they scored as well as every statistic associated with their shooting and passing to gain a better understanding of how they win basketball games.

```python
%%sql
CREATE OR REPLACE VIEW hometeam_stats AS
	SELECT game_details.Team_city,
	games.PTS_home,
	games.FG_PCT_home,
	games.FG3_PCT_home,
	games.AST_home,
	games.REB_home
	FROM games
	JOIN game_details
		ON games.Game_id = game_details.Game_id
	WHERE Home_team_wins = 1
	GROUP BY Team_city;


SELECT *
FROM hometeam_stats;
```

For my final data request I wanted to see all the players that scored more than the league average in pts. This will help me guauge which players are consistently high scorers and I can look down the list to find out who is an extreme outlier, this with a combination of plusminus will lead me to my answer. 

```python
%%sql
WITH player_rank AS (
	SELECT DISTINCT player.Player_name,
		player.Player_id,
		game_details.Team_city,
		SUM(game_details.PTS) OVER (
			PARTITION BY player.Player_name
		) AS total_pts
	FROM player 
	JOIN game_details
		ON player.Player_id = game_details.Player_id 
)
SELECT *
FROM player_rank
WHERE total_pts > (
	SELECT AVG(total_pts)
	FROM player_rank
);
```
